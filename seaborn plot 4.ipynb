```python
#Regression Plots
```

```python
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
df.head()
```

```python
#lmplot():
```

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',palette='rainbow')
```

```python
#Working with Markers:
# http://matplotlib.org/api/markers_api.html
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

```python
#Using a Grid:
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='smoker',row='time')
```

```python
#Aspect and Size:
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,height=8)
```
