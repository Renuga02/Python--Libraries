```python
# Seaborn Distribution Plots:
#1.distplot
#2.jointplot
#3.pairplot
#4.rugplot
#5.kdeplot
```

```python
import seaborn as sns
%matplotlib inline
```

```python
# load Defaulit dataset:
df=sns.load_dataset("tips")
```

```python
df
```

```python
df.info()
```

```python
df.shape
```

```python
#distplot:
```

```python
sns.displot(df['tip'])            # displot or  univariate
```

```python
sns.displot(df['total_bill'])
```

```python
sns.displot(df['total_bill'],kde=False,)
```

```python
#jointplot:
#types:
#“scatter”
#“reg”
#“resid”
#“kde”
#“hex”
```

```python
sns.jointplot(x='total_bill',y='tip',data=df)              # joint (scatter)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind= 'hex' )            # jointplot(hex)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')    #jointplot(reg)
```

```python
#pairplot:
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(data=df,hue='sex')
```

```python
#rugplot:
```

```python
sns.rugplot(df['total_bill'])               #rugplot
```

```python
#kdeplot (Kernel Density Estimation plots):
```

```python
sns.kdeplot(df['total_bill'])                     # kde plot
```

```python
sns.rugplot(df['total_bill'])
sns.kdeplot(df['total_bill'])
```
