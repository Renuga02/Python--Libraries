```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
#Styles:
```

```python
sns.countplot(x='time',data=df)
```

```python
#set style :
```

```python
sns.set_style('dark')
sns.countplot(x='time',data=df)
```

```python
#Spine Removal:
```

```python
sns.set_style('white')
sns.countplot(x='time',data=df,)
sns.despine()
```

```python
sns.set_style('white')
sns.countplot(x='time',data=df,)
sns.despine(left=True)
```

```python
#Size and Aspect:
```

```python
# Non Grid Plot:
```

```python
plt.figure(figsize=(8,2))
sns.countplot(x='sex',data=df)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',height=2,aspect=4,data=df)
```

```python
#Scale and Context:
```

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='time',data=df,palette='rainbow')
```
