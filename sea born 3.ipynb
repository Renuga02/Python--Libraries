```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Grids ( three thyes og grid)
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
df['species'].unique()
```

```python
df.head()
```

```python
# 1.pair grid:
# Just the Grid
```

```python
sns.PairGrid(df)
```

```python
# Then you map to the grid
g=sns.PairGrid(df)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
```

```python
g=sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#Pairplot:
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='species',palette='rainbow')
```

```python
#facet Grid:
```

```python
df1=sns.load_dataset('tips')
```

```python
df1
```

```python
df1.head()
```

```python

# Just the Grid
```

```python
sns.FacetGrid(data=df1,col='sex',row='smoker')
```

```python
f=sns.FacetGrid(data=df1,col='sex',row='smoker')
```

```python
f=sns.FacetGrid(data=df1,col='sex',row='smoker')
f.map(plt.hist,'total_bill')
```

```python
# Notice hwo the arguments come after plt.scatter call
```

```python
f=sns.FacetGrid(data=df1,col='sex',row='smoker',hue='time')
f.map(plt.scatter,'total_bill','tip').add_legend()
```

```python
#Join Grid:
```

```python
sns.JointGrid(x='total_bill',y='tip',data=df1)
```

```python
j=sns.JointGrid(x='total_bill',y='tip',data=df1)
j.plot(sns.regplot,sns.distplot)
```
