```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#The data we want to plot:
x=np.linspace(0,5,11)
y=x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')                                           #  create simple x,y plot 
plt.xlabel('x Axis')                                       # 'r' is the color red  
plt.ylabel('y Axis')
plt.title('Example')
```

```python
# Creating Multiplots on Same Canvas:
## plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,)
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#Introduction to the Object Oriented Method:

```

```python
# Create Figure (empty canvas)
fig=plt.figure()
# Add set of axes to figure
axes=fig.add_axes([0.1,0.1,.8,.8])      # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel(' X axis ')
axes.set_ylabel(' Y axis')
axes.set_title(' OOPs CONCEPT')
```

```python
# Create Figure (empty canvas)
fig=plt.figure()
# Add set of axes to figure
axes1=fig.add_axes([0.1,0.1,.8,.8]) 
axes2=fig.add_axes([0.2,0.2,0.4,.4])

axes1.plot(x,y,'b')
axes1.set_xlabel('set X axis')
axes1.set_ylabel('set Y axis')
axes1.set_title(' fig one')

axes2.plot(x,y,'g')
axes2.set_xlabel('set X axis')
axes2.set_ylabel(' set Y axis')
axes2.set_title(' fig two')

```

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig,axes=plt.subplots()

#Now use the axes object to add stuff to plot

axes.plot(x,y,'r')
axes.set_xlabel(' X axis')
axes.set_ylabel(' Y axis')
axes.set_title('SET TITLE')
```
