```python
#Categorical Data Plots:
# types:
#1.factorplot
#2.boxplot
#3.violinplot
#4.stripplot
#5.swarmplot
#6.barplot
#7.countplot
```

```python
import seaborn as sns
%matplotlib inline
```

```python
# load dataset:
df=sns.load_dataset('tips')
```

```python
df
```

```python
#barplot and countplot:
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='day',y='total_bill',data=df)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.std)
```

```python
#countplot:
```

```python
sns.countplot(x='day',data=df)
```

```python
sns.countplot(x='sex',data=df)
```

```python
#boxplot:
```

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
# Can do entire dataframe
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(data=df,palette='rainbow',orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=df)
```

```python
#violinplot:
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='sex',split='ture')
```

```python
#stripplot :
```

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.stripplot(x='day',y='total_bill',data=df,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=df,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=df,dodge=True,hue='sex')
```

```python
#swarmplot:
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df,hue='smoker',dodge=True)
```

```python
#Combining Categorical Plots:
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
sns.swarmplot(x='day',y='total_bill',data=df,color='black')
```

```python
#Catplot:
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='bar')
```

```python
sns.catplot(x='sex',y='total_bill',data=df)
```
